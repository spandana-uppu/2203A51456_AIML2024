```python
#Performance of Mathematical Operations on the given datasets Multiplication,Division,Poweroff,subtraction,modulo2 division
import numpy as np
dataset1=np.array([4,8,85,19])
dataset2=np.array([1,0.01,0.001,0.0001  ])
product=np.multiply(dataset1,dataset2)
product1=dataset1*dataset2
print('the product is as follows',product)
print('the product is as follows',product1)
Division=np.divide(dataset1,dataset2)
Moddivision=np.mod(dataset1,dataset2)
Floordivision=np.floor_divide(dataset1,dataset2)
Subtraction=np.subtract(dataset1,dataset2)
print(Division)
print(Moddivision)
print(Floordivision)
print(Subtraction)




```

```python
#implement Trigonometric and log functions of a given data set
import numpy as np
Patientage=np.array([65,78,45,67,56,77])
sin1=np.sin(Patientage)
cosine=np.cos(Patientage)
tangent=np.tan(Patientage)
Log1=np.log10(Patientage)
LOG=np.log2(Patientage)
EXP=np.exp(Patientage)
print('The sin of the patient age is ',sin1)
print('The cosine of the patient age is ',cosine)
print('The tangent of the patient age is ',tangent)
print('The log of the patient age is ',Log1)
print('The log base 2 of the patient age is ',LOG)
print('The exponential/antilogarithmic of the patient age is ',EXP)
```

Part-2:Lab Task is to visualize the datasets or data using Matplotlib function graphs


```python
people=['Dhanusri','Pranathi','shahana','Laxmi','Sreeja']
age=[17,19,20,21,24]
weight=[43,39,45,34,54]
height=[165,145,176,165,156]

import matplotlib.pyplot as plt
plt.scatter(weight,height)
plt.title("Relationship between weights and heights of People")
plt.ylabel("Height in meters")
plt.xlabel("Weigth in kgs")
plt.show()

```

Generate Bar Plot


```python
plt.figure(figsize=(12,4))
plt.title("People's weight in kgs",fontsize=16,fontstyle='italic')
#Main plot function 'bar'
plt.bar(x=people,height=weight,width=0.6,color='lavender',edgecolor='k',alpha=0.6)
plt.xlabel("People",fontsize=15)
plt.xticks(fontsize=14,rotation=30)
plt.yticks(fontsize=14)
plt.ylabel("Weight (in kgs)",fontsize=15)
plt.show()
```

Generate Histogram

```python
import numpy as np
plt.figure(figsize=(7,5))
#Main plot function 'hist'
plt.hist(weight,color='black',edgecolor='k',alpha=0.75,bins=5)
plt.title("Histogram of patient weight",fontsize=18)
plt.xlabel("Weight (in kgs)",fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.show()
```
